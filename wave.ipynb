```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import Audio


# Sample rate
fs = 20000
Ts = 1/fs
Tmax = 0.5
t = np.arange(0, Tmax, Ts)  # Time vector start/stop/step

# Wave
fb = 200
ub = np.cos(2 * np.pi * fb * t)
    
# Plot Time Domain
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st row
plt.title('time domain')
plt.plot(t, ub)
# plt.xlim(0, 0.1) # zoom if needed
plt.grid(True)
plt.xlabel('time')
plt.ylabel('voltage')
#plt.xlim(0, 0.02)  # Limit x-axis from 0 to 0.01

# Calculate Spectrum
F = np.fft.fft(ub)
n = len(ub)  # number of samples
f = np.fft.fftfreq(n, 1/fs) # frequency range
volt = 2/n * np.abs(F)  # voltage of the DFT

# Plot Frequency Domain
plt.subplot(2, 1, 2)  # 2 rows, 1 column, 2nd row
plt.title('frequency domain')
plt.plot(f[:n//2], volt[:n//2]) # plot only positive frequencies
plt.grid(True)    
plt.xlabel('frequency')
plt.ylabel('voltage')
plt.tight_layout()
plt.show()
    
# Audio Player
Audio(data = ub, rate = fs)
```
