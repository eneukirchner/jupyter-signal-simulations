```python
import numpy as np
import matplotlib.pyplot as plt

# Sample rate
fs = 10000
Ts = 1/fs
t = np.arange(0, 1, Ts)  # Time vector start/step/stop

# Generate wave
fb = 500
ub = np.cos(2 * np.pi * fb * t)

# Plot Time Domain
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st row
plt.plot(t, ub)
plt.grid(True)
plt.xlabel('time')
plt.ylabel('voltage')
plt.xlim(0, 0.01)  # Limit x-axis from 0 to 0.01

# Calculate Spectrum
F = np.fft.fft(ub)
n = len(ub)  # number of samples
f = np.arange(0, n) * (fs/n)  # frequency range
volt = 2/n * np.abs(F)  # voltage of the DFT
power = volt**2

# Plot Frequency Domain
plt.subplot(2, 1, 2)  # 2 rows, 1 column, 2nd row
plt.plot(f, power)
plt.grid(True)
plt.xlabel('frequency')
plt.ylabel('power')

plt.tight_layout()
plt.show()
```
